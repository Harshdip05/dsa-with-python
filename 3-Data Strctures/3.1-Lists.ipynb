#### Introduction To Lists
- Lists are ordered, mutable collections of items.
- They can contain items of different data types.

```python
lst = []
print(lst)
print(type(lst))
```

```python
names = ["k","h","a",1,2,3]
print(names)
```

```python
mixed_list = ["hello",1,2.4,True]
print(mixed_list)
```

```python
## Accessing list elements
fruits = ["banana","apple","kiwi","guava","watermelon"]
print(fruits)
print(len(fruits))

# list[start : stop : step]

# in R lang indexes start from 1 
```

```python
print(fruits[0])
print(fruits[-1])
```

```python
print(fruits[1:len(fruits)])
print(fruits[1:5])
print(fruits[1:])
```

```python
print(fruits[-1])
print(fruits[-1:])
print(fruits[-1:2])

print(fruits[-1:-3])  # By default, slicing uses a step of +1
print(fruits[-1:-3:-1])

print(fruits[-2:])
print(fruits[-2:-1])

print(fruits[-1:-10:-2])

print(fruits[::-1])
```

```python
## modifying list elements
fruits[0] = "mango"
print(fruits)
```

```python
fruits[1:] = "coconut"
print(fruits)
```

```python
## list methods

fruits = ["banana","apple","kiwi","guava","watermelon"]

fruits.append("custard")  # add an item to the end
print(fruits)
```

```python
fruits.insert(1,"banana")
print(fruits)
```

```python
fruits.remove("banana")  # remove first occurence 
print(fruits)
```

```python
# remove and return last element
pop_fruits = fruits.pop()
print(pop_fruits)
print(fruits)
```

```python
ind = fruits.index("cherry")
print(ind)
```

```python
ind1 = fruits.index("kiwi")
print(ind1)
```

```python
fruits.insert(1,"banana")

print(fruits.count("banana"))
print(fruits)
```

```python
fruits.sort()  # sort  list in ascending order
print(fruits)
```

```python
fruits.reverse()  # reverse list
print(fruits)
```

```python
fruits.clear()  # remove all items from list
print(fruits)
```

##### List Comprehension

Basics Syantax            [expression for item in iterable]

#### Conclusion
List comprehensions are a powerful and concise way to create lists in Python. They are syntactically compact and can replace more verbose looping constructs. Understanding the syntax of list comprehensions will help you write cleaner and more efficient Python code.
