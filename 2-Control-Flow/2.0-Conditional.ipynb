#### Conditional Statements (if, elif, else)

```python
## if statement

age = 20
if age >= 18 :
    print("you are allowed to vote in election")
```

```python
age >= 18
```

```python
## else
# else gets executed when if statement is False

age = 5
if age >= 18 :
    print("you are allowed to vote in election")
else:
    print("you are not able to vote")    
```

```python
## elif
# it allows to check for multiple conditions

age = 17
if age<13:
    print("you are child")
elif age < 18 :
    print("you are teenager")   
else:
    print("you are adult")    
```

```python
## Nested conditional statement
# you can place one or more if elif else inside another if elif else

num = int(input("enter number: "))
print(num)

if num> 0:
    print("number is positive")
    if num%2==0:
        print("even number")
    else:
        print("number is odd")    
else:
    print("number is zero or negative")
```

```python
## practical example

# determine if year is leap year or not using nesting

## The correct leap year rules are:
        # A year divisible by 400 → ✅ leap year
        # A year divisible by 100 but not by 400 → ❌ not a leap year
        # A year divisible by 4 but not by 100 → ✅ leap year
        # Otherwise → ❌ not a leap year

leap_year = int(input("enter year: "))
print(leap_year)

if leap_year%4==0:
    if leap_year%100==0:
        if leap_year%400==0:
             print("it is leap year")
        else:
             print("not a leap year")    
    else:
        print("is a leap year")    
else:
    print("not a leap year")
```

```python
year = int(input("Enter year: "))
print(year)

if year % 400 == 0:
    print("It is a leap year")
elif year % 100 == 0:
    print("Not a leap year")
elif year % 4 == 0:
    print("It is a leap year")
else:
    print("Not a leap year")
```

```python
year = int(input("Enter year: "))
print(year)

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("It is a leap year")
else:
    print("Not a leap year")
```

```python
## simple calculator

num1 = float(input("enter number: "))
num2 = float(input("enter number: "))
print(num1,num2)

ops = input("enter operation: ")

if ops == "+":
    print("addition",num1+num2)
elif ops == "-":
    print("subtraction",num1-num2)
elif ops == "*":
    print("multiplication",num1*num2)
elif ops == "/":
    if num2!=0:
        print("division",num1/num2)
    else:
        print("error, division by zero")    
else:
    print("wrong entries")    
```

```python
## Assignment
## Simple Calculator program
# Take user input
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
print(num1,num2)
operation = input("Enter operation (+, -, *, /): ")

# Perform the requested operation
if operation == '+':
    result = num1 + num2
elif operation == '-':
    result = num1 - num2
elif operation == '*':
    result = num1 * num2
elif operation == '/':
    if num2 != 0:
        result = num1 / num2
    else:
        result = "Error! Division by zero."
else:
    result = "Invalid operation."

print("Result:", result)
```

```python
## determine ticket price based on age and whether a person is a student
# Ticket pricing based on age and student status

# Take user input
age = int(input("Enter your age: "))
is_student = input("Are you a student? (yes/no): ").lower()

# Determine ticket price
if age < 5:
    price = "Free"
elif age <= 12:
    price = "$10"
elif age <= 17:
    if is_student == 'yes':
        price = "$12"
    else:
        price = "$15"
elif age <= 64:
    if is_student == 'yes':
        price = "$18"
    else:
        price = "$25"
else:
    price = "$20"

print("Ticket Price:", price)
```

```python
# Employee bonus calculation

# Take user input
years_of_service = int(input("Enter years of service: "))
performance_rating = float(input("Enter performance rating (1.0 to 5.0): "))

# Determine bonus percentage
if performance_rating >= 4.5:
    if years_of_service > 10:
        bonus_percentage = 20
    elif years_of_service > 5:
        bonus_percentage = 15
    else:
        bonus_percentage = 10
elif performance_rating >= 3.5:
    if years_of_service > 10:
        bonus_percentage = 15
    elif years_of_service > 5:
        bonus_percentage = 10
    else:
        bonus_percentage = 5
else:
    bonus_percentage = 0

# Calculate bonus amount
salary = float(input("Enter current salary: "))
bonus_amount = salary * bonus_percentage / 100

print("Bonus Amount: ${:.2f}".format(bonus_amount))
```

```python
## Complex Example 4: User Login System
# A simple user login system that checks the username and password.

# User login system

# Predefined username and password
stored_username = "admin"
stored_password = "password123"

# Take user input
username = input("Enter username: ")
password = input("Enter password: ")

# Check login credentials
if username == stored_username:
    if password == stored_password:
        print("Login successful!")
    else:
        print("Incorrect password.")
else:
    print("Username not found.")
```
