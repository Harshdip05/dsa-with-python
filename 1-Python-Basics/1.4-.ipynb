An escape sequence is a special character inside a string that starts with a backslash \.

It tells Python:

“Don’t treat this normally — do something special.”

\n  
\t  
\'  
\"  
\\  
\r  
\b

```python
# \n → New Line
print("Hello\nWorld")

# \t → Tab
print("Hello\tWorld")

# \\ → Backslash
print("C:\\Users\\Admin")

# \' → Single Quote
print('It\'s Python')

# \" → Double Quote
print("He said \"Hello\"")

# \r → Carriage Return
print("Hello\rWorld")

# \b → Backspace
print("Helloo\b")

# Raw String (ignores escape sequences)
print(r"C:\Users\Admin")

# Example showing problem without raw string
print("C:\\newfolder")   # Correct way
```
